```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/home/mzo5929/Keerthana/grnInference/simulationData/aGuptaSimulation_20250521/samples_replicates_with_regulation_no_burn_in_median_0.csv")
```

```python
df = pd.read_csv("/home/mzo5929/Keerthana/grnInference/simulationData/aGuptaSimulation_20250521/samples_replicates_with_regulation_no_burn_in_median_1.csv")
```

```python
#Plot correlation between total_TF_mRNA and total_Target_mRNA at each time point
corr_list = []
for time in df['sampling_time'].unique():
    temp_df = df[df['sampling_time'] == time]
    correlation = temp_df['total_TF_mRNA'].corr(temp_df['total_Target_mRNA'])
    protein_level = temp_df['TF_protein_1K'].mean()  # Example of using protein level, if needed
    protein_level_std = temp_df['TF_protein_1K'].std()  # Example of using protein level std, if needed
    corr_list.append((time, correlation, protein_level, protein_level_std))
times, correlations, protein_level, protein_level_std = zip(*corr_list)
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(times, correlations, marker='o')
plt.title('Correlation between total TF mRNA and total Target mRNA over time')
plt.xlabel('Time')
plt.ylabel('Correlation Coefficient')
plt.grid()
plt.tight_layout()
# plt.savefig('correlation_TF_Target_over_time.png')
plt.show()

#Plot protein level distributions over tine
plt.figure(figsize=(10, 6))
plt.plot(times, protein_level, label='Mean TF Protein Level')
plt.plot(times, protein_level_std, label='TF Protein Level Std Dev')
plt.title('TF Protein Level Distribution over Time')
plt.xlabel('Time')
plt.ylabel('TF Protein Level (1K)')
plt.legend()
plt.show()
```
